```python
import matplotlib.pyplot as plt

from PIL import Image
src = Image.open("taichi.jpg")#画像をセット
plt.imshow(src)
plt.show()
```

```python
#45度に傾ける
dest = src.rotate(45)

plt.imshow(dest)
plt.show()
#dest.save("dest.jpg","JPEG")
```

```python
#画像のサイズを100x100ピクセルにリサイズする
resized = src.resize((100,100))
plt.imshow(resized)
plt.show()
# resized.save("resized.jpg","JPEG")
```

```python
#画像を3色に減色させる
#減色をした画像を、パレットを持たないJPEGのような形式で保存することはできない
#パレットを持つPNGやBMP形式で保存しよう
down_color = src.quantize(3)

plt.imshow(down_color)
plt.show()
# down_color.save("down_color.png","PNG")
```

```python
#ブラー効果(ぼやける)を加える
from PIL import ImageFilter
blur = src.filter(ImageFilter.BLUR)

plt.imshow(blur)
plt.show()
#blur.save("blur.jpg","JPEG")
```

```python
#輪郭効果を加える
contour = src.filter(ImageFilter.CONTOUR)

plt.imshow(contour)
plt.show()
# contour.save("contour.jpg","JPEG")
```

```python
#画像のエッジを残したまま高画質化
detail = src.filter(ImageFilter.DETAIL)

plt.imshow(detail)
plt.show()
#detail.save("detail.jpg","JPEG")
```

```python
#輪郭を強調
edge = src.filter(ImageFilter.EDGE_ENHANCE)

plt.imshow(edge)
plt.show()
# edge.save("edge.jpg","JPEG")
```

```python
#より輪郭を強調
edge_more = src.filter(ImageFilter.EDGE_ENHANCE_MORE)

plt.imshow(edge_more)
plt.show()
# edge_more.save("edge_more.jpg","JPEG")
```

```python
#エンボス
emboss = src.filter(ImageFilter.EMBOSS)

plt.imshow(emboss)
plt.show()
#emboss.save("emboss.jpg","JPEG")
```

```python
#輪郭抽出
find_edges = src.filter(ImageFilter.FIND_EDGES)

plt.imshow(find_edges)
plt.show()
#find_edges.save("find_edges.jpg","JPEG")
```

```python
#スムージング
smooth = src.filter(ImageFilter.SMOOTH)

plt.imshow(smooth)
plt.show()
#smooth.save("smooth.jpg","JPEG")
```

```python
#シャープネス
sharpen = src.filter(ImageFilter.SHARPEN)

plt.imshow(sharpen)
plt.show()
# sharpen.save("sharpen.jpg","JPEG")
```
